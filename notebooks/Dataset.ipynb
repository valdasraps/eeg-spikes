```python
import pandas as pd
import numpy as np
```

```python
pdata = pd.read_csv('../data/spike-params-master7.csv')
```

```python
pdata
```

```python
meta = pdata[['PID','PEAK','CH','DIAG']].values.astype(str)
meta = np.char.strip(meta)
meta.shape
```

```python
data = pdata[list(map(str,range(1,78)))].values.astype(float)
data.shape
```

```python
meta[np.unique(np.where(np.isnan(data))[0]),:2]
```

```python
groups = np.unique(meta[:,[0,1,3]], axis = 0)
groups.shape
```

```python
channels = np.sort(np.unique(meta[:,2]))
channels.shape
```

```python
data_groups = np.zeros((len(meta)), dtype=int)
data_groups.shape
```

```python
for i, g in enumerate(groups):
    a = meta[:,[0,1]] == g[:2]
    data_groups[a[:,0] & a[:,1]] = i
```

```python
matrix = np.zeros((len(groups),len(channels),data.shape[1]), dtype = float)
matrix.shape
```

```python
for i, g in enumerate(groups):
    try:
        mask = data_groups == i
        ch = meta[mask][:,2]
        _, uindex = np.unique(ch, return_index = True)
        if np.all(np.isin(ch[uindex], channels, assume_unique = True)):
            chorder = np.argsort(ch[uindex])
            matrix[i,:,:] = data[mask][uindex][chorder]
        else:
            raise Exception(f'{g}')
    except Exception as e:
        print(f'{g}: {e}')
        break
```

```python
np.save('../data/data.npy', matrix)
np.save('../data/meta.npy', groups.astype(float))
```
